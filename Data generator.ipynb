### Data generator

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
for i in range(6):
    fname = f"position/position-{i:02d}.dat"
    n_clusters = np.random.randint(6,13)
    X, y = make_blobs(n_samples=10000, cluster_std=120, centers=n_clusters, n_features=2, center_box=(-500.0, 500.0))
    pickle.dump(X, open(fname, "wb"))
```

```python
for i in range(6):
    fname = f"position/position-{i:02d}.dat"
    X = pickle.load(open(fname, "rb"))
    plt.scatter(X[:,0], X[:,1], s=2)
    plt.xlim((-800,800))
    plt.ylim((-800,800))
    plt.show()
```
